```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import torch
import os

# RUTA BASE
data_dir = r"C:\Users\soria\Downloads\Metodología\Clasificacion_residuos"

# TRANSFORMACIONES
# Entrenamiento: incluye aumentos para robustez del modelo
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# Validación / prueba: solo escalado y normalización
test_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# CARGA DE DATOS (70/30 reproducible)
def cargar_dataset(folder_name, split_ratio=0.7, random_seed=42):
    """
    Carga un dataset (TrashNet o TACO) y lo divide en 70/30 para
    entrenamiento y validación. No modifica las carpetas originales.
    """
    torch.manual_seed(random_seed)  # semilla fija para reproducibilidad

    # Dataset completo con transformaciones de entrenamiento
    full_dataset = datasets.ImageFolder(os.path.join(data_dir, folder_name),
                                        transform=train_transforms)

    dataset_size = len(full_dataset)
    train_size = int(split_ratio * dataset_size)
    val_size = dataset_size - train_size

    # División reproducible 70/30
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    dataloaders = {
        'train': DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2),
        'val': DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
    }

    dataset_sizes = {'train': len(train_dataset), 'val': len(val_dataset)}
    class_names = full_dataset.classes

    print(f" Clases detectadas: {class_names}")
    print(f" Imágenes en entrenamiento: {dataset_sizes['train']}")
    print(f" Imágenes en validación: {dataset_sizes['val']}")

    return dataloaders, dataset_sizes, class_names

# PRUEBA CON TRASHNET
dataloaders, dataset_sizes, class_names = cargar_dataset("TACO")

```

```python
import torch.nn as nn
import torch.optim as optim
from torchvision import models
import copy
import time

# === Definición del modelo ResNet-18 ===
num_classes = len(class_names)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Cargar la ResNet-18 preentrenada en ImageNet
model = models.resnet18(weights="IMAGENET1K_V1")

# Congelar las capas convolucionales (fase 1)
for param in model.parameters():
    param.requires_grad = False

# Reemplazar la capa final (fc) con una nueva que tenga 6 salidas
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, num_classes)
model = model.to(device)

# Definir la función de pérdida y el optimizador
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=0.001)

print(" Modelo ResNet-18 preparado para entrenamiento.")
print(f"Número de clases: {num_classes}")
print(f"Dispositivo en uso: {device}")
```

```python
import os
import time
import copy
import torch.nn.functional as F

# Crear carpeta "Modelos" si no existe
os.makedirs("Modelos", exist_ok=True)

def train_model(model, criterion, optimizer, dataloaders, dataset_sizes, escenario="S0", num_epochs=10):
    """
    Entrena un modelo ResNet-18 y guarda el archivo .pth con el nombre del escenario.
    """
    since = time.time()
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    print(f"\n Iniciando entrenamiento para el escenario {escenario}")
    print("-" * 50)

    for epoch in range(num_epochs):
        print(f"\n Época {epoch+1}/{num_epochs}")
        print('-' * 30)

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs, labels = inputs.to(device), labels.to(device)
                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print(f'{phase.upper():<6} | Pérdida: {epoch_loss:.4f} | Exactitud: {epoch_acc:.4f}')

            # Guardar el mejor modelo
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())

    time_elapsed = time.time() - since
    print(f"\n Entrenamiento completo en {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s")
    print(f" Mejor exactitud de validación: {best_acc:.4f}")

    # Cargar los pesos del mejor modelo
    model.load_state_dict(best_model_wts)

    # Guardar el modelo
    model_path = f"Modelos/resnet18_{escenario}.pth"
    torch.save(model.state_dict(), model_path)
    print(f"\n Modelo guardado como: {model_path}")

    return model


# FASE 1: Entrenamiento solo de la capa final (capas congeladas)
model = train_model(model, criterion, optimizer, dataloaders, dataset_sizes, escenario="S0", num_epochs=10)

# FASE 2: Fine-Tuning: descongelar todas las capas 
for param in model.parameters():
    param.requires_grad = True

optimizer = optim.Adam(model.parameters(), lr=1e-4)
model = train_model(model, criterion, optimizer, dataloaders, dataset_sizes, escenario="S0_finetuned", num_epochs=10)
```

```python
import torch
import torch.nn as nn
from torchvision import models, datasets, transforms
from torch.utils.data import DataLoader
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import os

# === Configuración ===
data_dir = r"C:\Users\soria\Downloads\Metodología\Clasificacion_residuos"

eval_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

def cargar_modelo(path_modelo, num_classes):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = models.resnet18(weights=None)
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, num_classes)
    model.load_state_dict(torch.load(path_modelo, map_location=device))
    model = model.to(device)
    model.eval()
    return model, device

def plot_confusion_matrix(cm, class_names, title):
    fig, ax = plt.subplots(figsize=(6,5))
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels=class_names, yticklabels=class_names,
           title=title,
           ylabel='Etiqueta Real',
           xlabel='Predicción')
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], 'd'),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    plt.show()

def evaluar_modelo(model, dataloader, class_names, device, escenario="Sx"):
    y_true, y_pred = [], []
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    acc = accuracy_score(y_true, y_pred)
    print(f"\n=== Resultados del Escenario {escenario} ===")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))
    print(f"🔹 Exactitud global: {acc:.4f}")

    cm = confusion_matrix(y_true, y_pred)
    plot_confusion_matrix(cm, class_names, f'Matriz de Confusión - {escenario}\nAccuracy: {acc:.2f}')
    return acc

# === Datasets ===
trashnet_eval = datasets.ImageFolder(os.path.join(data_dir, "TrashNet"), transform=eval_transform)
taco_eval = datasets.ImageFolder(os.path.join(data_dir, "TACO"), transform=eval_transform)
dataloader_trashnet = DataLoader(trashnet_eval, batch_size=32, shuffle=False)
dataloader_taco = DataLoader(taco_eval, batch_size=32, shuffle=False)

class_names = trashnet_eval.classes
num_classes = len(class_names)

# === Modelos ===
modelos = {
    "S0_finetuned": ("Modelos/resnet18_S0_finetuned.pth", dataloader_trashnet),
    "S1_finetuned": ("Modelos/resnet18_S1_finetuned.pth", dataloader_taco),
    "S2_finetuned": ("Modelos/resnet18_S2_finetuned.pth", dataloader_taco),
    "S3_finetuned": ("Modelos/resnet18_S3_finetuned.pth", dataloader_trashnet)
}

accuracy_dict = {}

for esc, (path, loader) in modelos.items():
    model, device = cargar_modelo(path, num_classes)
    acc = evaluar_modelo(model, loader, class_names, device, escenario=esc)
    accuracy_dict[esc] = acc

# === Comparativa final ===
plt.figure(figsize=(8,5))
escenarios = list(accuracy_dict.keys())
accuracies = [accuracy_dict[e] for e in escenarios]
bars = plt.bar(escenarios, accuracies, color=['#1f77b4','#2ca02c','#d62728','#9467bd'])
plt.ylim(0, 1)
plt.title("Comparación de Exactitud por Escenario (ResNet-18)")
plt.xlabel("Escenario")
plt.ylabel("Exactitud (Accuracy)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f"{yval:.2f}",
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n===== RESUMEN FINAL DE ACCURACY =====")
for k, v in accuracy_dict.items():
    print(f"{k}: {v:.4f}")
```
